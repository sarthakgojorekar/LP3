```python
import numpy as np
def calculate_dist(lon1, lat1, lon2, lat2):
    delta_lon = lon1 - lon2
    delta_lat = lat1 - lat2
    dist = np.sqrt(delta_lon**2 + delta_lat**2)
    return dist
```
