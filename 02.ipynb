```python
import pandas as pd 
```

```python
df = pd.read_csv('./Churn_Modelling.csv')
df
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

le = LabelEncoder()

df['Gender'] = le.fit_transform(df['Gender'])
df['Geography'] = le.fit_transform(df['Geography'])
df
```

```python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'],axis= 1)
df
```

```python
x = df.drop('Exited', axis = 1)
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
sc = StandardScaler()

x_test = sc.fit_transform(x_test)
x_train = sc.fit_transform(x_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(activation = 'relu', units = 32, input_dim = x_train.shape[0]))
model.add(Dropout(0.5))
model.add(Dense(activation = "relu", units = 16))
model.add(Dense(activation = "sigmoid", units = 1))

model.complie(optimiser = 'adam', loss = 'cross-entropy',metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, batch_size = 10, random_state = 43)
```

```python
y_pred = model.predit(x_test)
```

```python
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
```
